```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("chennai-properties-cleaned.xlsx")
```

```python
df["price_per_sqft"] = (df["price_lakhs"] * 100000) / df["area_sqft"]
```

```python
# 1. Price Distribution
plt.figure(figsize=(10, 5))
sns.histplot(df["price_lakhs"], bins=30, kde=True)
plt.title("Distribution of Property Prices (Lakhs)")
plt.xlabel("Price (Lakhs)")
plt.ylabel("Count")
plt.grid()
plt.show()
```

```python
# 2. Top 10 Localities by Listings
top_locations = df["location"].value_counts().head(10)
plt.figure(figsize=(10, 5))
sns.barplot(x=top_locations.index, y=top_locations.values, palette="viridis")
plt.title("Top 10 Localities by Number of Listings")
plt.xlabel("Location")
plt.ylabel("Number of Listings")
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
# 3. BHK Distribution
plt.figure(figsize=(6, 6))
df["bhk"].value_counts().sort_index().plot.pie(autopct="%1.1f%%", startangle=90, colors=sns.color_palette("Set3"))
plt.title("BHK Type Distribution")
plt.ylabel("")
plt.show()
```

```python
# 4. Area vs Price
plt.figure(figsize=(10, 5))
sns.scatterplot(data=df, x="area_sqft", y="price_lakhs", hue="bhk", palette="deep")
plt.title("Area vs Price")
plt.xlabel("Area (sqft)")
plt.ylabel("Price (Lakhs)")
plt.grid()
plt.show()
```

```python
# 5. Area vs Price per Sqft
plt.figure(figsize=(10, 5))
sns.scatterplot(data=df, x="area_sqft", y="price_per_sqft", hue="bhk", palette="coolwarm")
plt.title("Area vs Price per Sqft")
plt.xlabel("Area (sqft)")
plt.ylabel("Price per Sqft (INR)")
plt.grid()
plt.show()
```
